# Extração de Seções de Artigos Científicos em PDF
Este notebook extrai automaticamente as seções **Resumo**, **Metodologia**, **Resultados** e **Conclusão** de um artigo científico em PDF usando a biblioteca `PyMuPDF`.

```python
# Instalar o PyMuPDF se necessário
!pip install pymupdf
```

```python
import fitz  # PyMuPDF
import re

def extrair_texto_pdf(caminho_pdf):
    doc = fitz.open(caminho_pdf)
    texto = "\n".join([pagina.get_text() for pagina in doc])
    return texto

def extrair_secao(texto, inicio, fim=None):
    padrao = rf"{inicio}(.*?){fim if fim else '$'}"
    resultado = re.search(padrao, texto, re.DOTALL | re.IGNORECASE)
    return resultado.group(1).strip() if resultado else "Seção não encontrada."
```

```python
# Substitua pelo caminho do seu arquivo PDF
caminho_pdf = "/content/25308-673-20783-2-10-20230803 (1).pdf"
texto = extrair_texto_pdf(caminho_pdf)

# Extrair as seções
resumo = extrair_secao(texto, r"RESUMO", r"(?:1 Introdução|1\. Introdução|INTRODUÇÃO)")
metodologia = extrair_secao(texto, r"3\. Materiais e Método", r"4\. Experimentos e Resultados")
resultados = extrair_secao(texto, r"4\. Experimentos e Resultados", r"5\. Conclusão")
conclusao = extrair_secao(texto, r"5\. Conclusão", r"6\. Agradecimentos")
```

```python
# Exibir os resultados extraídos
print("=== RESUMO ===\n", resumo, "\n")
print("=== METODOLOGIA ===\n", metodologia, "\n")
print("=== RESULTADOS ===\n", resultados, "\n")
print("=== CONCLUSÃO ===\n", conclusao, "\n")
```
